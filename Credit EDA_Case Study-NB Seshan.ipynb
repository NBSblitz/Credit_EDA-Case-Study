# Data Cleaning

```python
df.isnull().sum().sort_values(ascending=False)
```

```python
#Converting the data type of column SK_ID_CURR to object
df['SK_ID_CURR']=df['SK_ID_CURR'].astype(object)
df.info()
```

```python
# Converting the data type of column 'CNT_CHILDREN' to int and 'REGION_RATING_CLIENT_W_CITY' to object
df['CNT_CHILDREN']=df['CNT_CHILDREN'].astype(int)
df['REGION_RATING_CLIENT_W_CITY']=df['REGION_RATING_CLIENT_W_CITY'].astype(object)
df.info()
```

```python
df.describe()
```

```python
# Identifying columns with negative values
numeric_columns=df.select_dtypes(include=['number'])  
negative_columns=numeric_columns.columns[(numeric_columns<0).any()]
negative_columns
```

```python
# Converting the negative columns to positive
df[negative_columns]=df[negative_columns].abs()
df.describe()
```

```python
# Adding two new columns 'CLIENT_AGE' and 'EMPLOYED_YEARS'
df['CLIENT_AGE']=df['DAYS_BIRTH']//365
df['EMPLOYED_YEARS']=df['DAYS_EMPLOYED']//365
df.describe()
```

```python
#Taking the values in the 'EMPLOYED_YEARS' column till the 80th percentile
q1=df['EMPLOYED_YEARS'].quantile(0.80)
df['EMPLOYED_YEARS']=df.EMPLOYED_YEARS.apply(lambda x: q1 if x > q1 else x)
```

```python
#Taking the values in the 'CNT_CHILDREN' column till the 99th percentile
q2=df['CNT_CHILDREN'].quantile(0.99)
df['CNT_CHILDREN']=df.CNT_CHILDREN.apply(lambda x: q2 if x>q2 else x)
```

```python
#Taking the values in the 'AMT_INCOME_TOTAL' column till the 95th percentile
q3=df['AMT_INCOME_TOTAL'].quantile(0.95)
df['AMT_INCOME_TOTAL']=df.AMT_INCOME_TOTAL.apply(lambda x: q3 if x>q3 else x)
df.head()
```

```python
# Binning salary of clients
def salary_category(x):
    if x>=225000:
        return('HIGH')
    elif x<225000 and x>=147600:
        return('MEDIUM')
    elif x<147600 and x>=112500:
        return('LOW')
    else:
        return("VERY LOW")
df['SALARY_CATEGORY']=df.AMT_INCOME_TOTAL.apply(salary_category)
df['SALARY_CATEGORY'].head()
```

```python
df['SALARY_CATEGORY'].value_counts()
```

```python
# Binning ages of clients
df['CLIENT_AGE_BAND'] = pd.cut(df['CLIENT_AGE'],[0,17,25,35,45,55,np.inf])
df.head()
```

### Dividing the dataframe into 2 datasets based on target variables

```python
good_clients=df[df.TARGET == 0]
defaulter_clients=df[df.TARGET==1]
```

```python
def plot_percent_defaulter(defaulter_clients,category_col,title,r):
    percent_defaulter = defaulter_clients[category_col].value_counts() / df[category_col].value_counts() * 100
    plt.figure(figsize=(8,6))
    sns.barplot(x=percent_defaulter.index,y=percent_defaulter.values,palette='Set2')
    plt.title(title,fontsize=15)
    plt.xlabel(category_col,fontsize=12)
    plt.ylabel('Percentage of Defaulting Clients',fontsize=12)
    plt.xticks(rotation=r)
    plt.show() 
```

```python
def plot_percent_good(good_clients,category_col,title,r):
    percent_good = good_clients[category_col].value_counts() / df[category_col].value_counts() * 100
    plt.figure(figsize=(8,6))
    sns.barplot(x=percent_good.index,y=percent_good.values,palette='Set2')
    plt.title(title,fontsize=15)
    plt.xlabel(category_col,fontsize=12)
    plt.ylabel('Percentage of Good Clients',fontsize=12)
    plt.xticks(rotation=r)
    plt.show() 
```

```python
# Plotting the percentage of defaulting clients for the column 'CODE_GENDER'
plot_percent_defaulter(defaulter_clients,'CODE_GENDER','Percentage of Defaulting Clients by Gender',0)
```
